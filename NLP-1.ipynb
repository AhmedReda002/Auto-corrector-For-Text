```python
def counting_words(words):
    words_count = {}
    for word in words:
        if word in words_count:
            words_count[word] += 1
        else:
            words_count[word] = 1
    return words_count


dic_words=counting_words(words)
```

```python
def prop_cal(dic):
    props = {}
    total = sum(dic.values())
    
    for key in dic.keys():
        props[key] = dic[key] / total
    return props

prop_cal(dic_words)
```

```python
def split_listfunc(word):
    split_list=[]
    for i in range(len(word)):
        split_list.append((word[:i], word[i:]))
        
    return split_list    
```

```python
def Deleteletter(word):
    delete_list = []
    split_list = split_listfunc(word)
    
    # Generate all splits of the word
    for i in range(len(word)):
        split_list.append((word[:i], word[i:]))
    
    # Generate all strings by deleting one letter at a time
    for p1, p2 in split_list:
        delete_list.append(p1 + p2[1:])  # Remove one character from p2
    
    return delete_list

# Example usage:
#result = Deleteletter('nice')
#print(result)
```

```python
def switch(word):
    split_list = split_listfunc(word)
    switch_list = []
    
    # Perform the switch operation
    for p1, p2 in split_list:
        if len(p2) >= 2:
            switch_list.append(p1 + p2[1] + p2[0] + p2[2:])
        else:
            switch_list.append(p1 + p2)  # Append the unchanged substring if length < 2
    
    return switch_list

# Example usage:
#result = switch('nice')
#print(result)
```

```python
def replace(word):
    def split_listfunc(word):
        split_list = []
        for index in range(len(word)):
            split_list.append((word[:index], word[index:]))
        return split_list
    
    split_list = split_listfunc(word)
    replace_list = []
    
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    
    for p1, p2 in split_list:
        for c in alphabet:
            if len(p2) > 0:  # Ensure there is at least one character in p2 to replace
                replace_list.append(p1 + c + p2[1:])
    
    return replace_list

# Example usage:
#result = replace('nace')
#print(result)
```

```python
def insert(word):
    def split_listfunc(word):
        split_list = []
        for index in range(len(word) + 1):  # Include one more position for the end of the word
            split_list.append((word[:index], word[index:]))
        return split_list
    
    split_list = split_listfunc(word)
    insert_list = []
    
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    
    for p1, p2 in split_list:
        for c in alphabet:
            insert_list.append(p1 + c + p2)
    
    return insert_list

# Example usage:
#result = insert('nice')
#print(result)
```
